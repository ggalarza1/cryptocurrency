```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# Dowloading historical data for XRP and fetching historical data with adjusted period
ticker = yf.Ticker("DOGE-USD")
ticker_hist = ticker.history(period="max")

# Extracting the closing price
doge = ticker_hist['Close']

# Graphing the historical data
fig, ax = plt.subplots()

ax.plot(doge.index, doge, label='Dogecoin')

# Adding the labels to the data
ax.set_xlabel('Date')
ax.set_ylabel('Closing price ($)')
ax.set_title(label= 'Dogecoin', loc='center')
ax.legend()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.dates as mdates
import datetime as dt

# Dowloading historical data for XRP and fetching historical data with adjusted period
ticker = yf.Ticker("DOGE-USD")
ticker_hist = ticker.history(period="max")

# Extracting the closing price
doge = ticker_hist['Close']

# Graphing the historical data
fig = plt.figure()
ax = plt.axes(projection='3d')

#Convert the DatetimeIndex to numerical values using matplotlib.dates.date2num
x_values = mdates.date2num(doge.index)

# defining surface and axes
doge_values = doge.to_numpy()

# Create a meshgrid for x and y using the length of xrp_values
num_points = len(doge_values)
x, y = np.meshgrid(np.linspace(-2, 2, num_points), np.linspace(-2, 2, num_points))


# Calculate z using the meshgrid x and y
z = np.cos(x ** 2 + y ** 3)  # or any function of x, y

# plotting
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='green')

ax.set_title('DOGE')
ax.set_ylabel('Closing price ($)')
ax.set_xlabel('Date')


plt.show()
```

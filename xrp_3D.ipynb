```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.dates as mdates
import datetime as dt

# Dowloading historical data for XRP and fetching historical data with adjusted period
ticker = yf.Ticker("XRP-USD")
ticker_hist = ticker.history(period="max")

# Extracting the closing price
xrp = ticker_hist['Close']

# Graphing the historical data
fig = plt.figure()
ax = plt.axes(projection='3d')

#Convert the DatetimeIndex to numerical values using matplotlib.dates.date2num
x_values = mdates.date2num(xrp.index)

# Plot the data using the numerical x-values
ax.plot(x_values, xrp, label='XRP')

# defining all 3 axis
z = xrp
x = np.array(range(len(xrp))) * np.sin(25 * np.array(range(len(xrp)))) # Use numerical values for x-axis
y = np.array(range(len(xrp))) * np.cos(25 * np.array(range(len(xrp))))# Use numerical values for y-axis

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('XRP')
ax.set_ylabel('Closing price ($)')
ax.set_xlabel('Date')

# Format the x-axis to display dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.dates as mdates
import datetime as dt

# Dowloading historical data for XRP and fetching historical data with adjusted period
ticker = yf.Ticker("XRP-USD")
ticker_hist = ticker.history(period="max")

# Extracting the closing price
xrp = ticker_hist['Close']

# Graphing the historical data
fig = plt.figure()
ax = plt.axes(projection='3d')

#Convert the DatetimeIndex to numerical values using matplotlib.dates.date2num
x_values = mdates.date2num(xrp.index)

# Plot the data using the numerical x-values
ax.plot(x_values, xrp, label='XRP')

# defining surface and axes
xrp_values = xrp.to_numpy()
x = np.outer(xrp_values, np.linspace(-2, 2, 10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)

# plotting
ax.plot_surface(x, y, z, cmap='viridis',\
                edgecolor='green')
#ax.plot3D(x, y, z, 'green')
ax.set_title('XRP')
ax.set_ylabel('Closing price ($)')
ax.set_xlabel('Date')

# Format the x-axis to display dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.show()
```
